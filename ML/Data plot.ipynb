```python
from sklearn.decomposition import PCA
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data_transformed.csv',delimiter=',')
row = data.shape[0]
col = data.shape[1]
labels = data[0:row,col-2:col]
data = data[0:row,0:col-2]
data = preprocessing.scale(data)
```

```python
pca = PCA(n_components = 2)
data = pca.fit_transform(data)
print(data.shape)
```

```python
datas = [[],[],[],[],[]]
for i in range(row):
    datas[int(labels[i,0])-1].append(data[i,:])
for i in range(5):
    datas[i] = np.array(datas[i])
    print(datas[i].shape)
```

```python
plt.figure(figsize=(10,10))
c1 = plt.scatter(x = datas[0][0:datas[0].shape[0],0],y = datas[0][0:datas[0].shape[0],1],c ='red')
c2 = plt.scatter(x = datas[1][0:datas[1].shape[0],0],y = datas[1][0:datas[0].shape[0],1],c ='green')
c3 = plt.scatter(x = datas[2][0:datas[2].shape[0],0],y = datas[2][0:datas[0].shape[0],1],c ='blue')
c4 = plt.scatter(x = datas[3][0:datas[3].shape[0],0],y = datas[3][0:datas[0].shape[0],1],c ='yellow')
c5 = plt.scatter(x = datas[4][0:datas[4].shape[0],0],y = datas[4][0:datas[0].shape[0],1],c ='purple')
plt.title('30to2-PCA Data Ditribution On Dalc')
plt.legend(handles=[c1, c2, c3, c4, c5], labels=['class1', 'class2', 'class3', 'class4', 'class5'])
plt.show()
```
