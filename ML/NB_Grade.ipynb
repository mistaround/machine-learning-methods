```python
import numpy as np
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
data = np.loadtxt('grade.csv',delimiter=',')
row = data.shape[0]
col = data.shape[1]

label = data[:,col-3]
data = data[:,:col-3]

for i in range(row):
    if label[i] <= 4:
        label[i] = 1
    elif label[i] <= 8:
        label[i] = 2
    elif label[i] <= 12:
        label[i] = 3
    elif label[i] <= 16:
        label[i] = 4
    elif label[i] <= 20:
        label[i] = 5
```

```python
train, test, train_label_1, test_label_1 = train_test_split(data, label, test_size=0.25)
train = preprocessing.scale(train)
test = preprocessing.scale(test)
train_row = train.shape[0]
test_row = test.shape[0]
```

```python
gnb = GaussianNB()
gnb.fit(train, train_label_1)
```

```python
c = 0
for i in range(train_row):
    if gnb.predict([train[i,:]]) != train_label_1[i]:
        print(i,gnb.predict([train[i,:]]),train_label_1[i])
        c += 1
print(1-c/train_row)
```

```python
c = 0
for i in range(test_row):
    if gnb.predict([test[i,:]]) != test_label_1[i]:
        print(i,gnb.predict([test[i,:]]),test_label_1[i])
        c += 1
print(1-c/test_row)
```
